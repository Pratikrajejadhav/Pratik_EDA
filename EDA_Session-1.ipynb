# create dataframe using List

## step-1 : **Create a Dataframe using List**

```python
Names = ['Ramesh','Suresh','Sathis']
Age = [20,25,30]
pd.DataFrame(zip(Names,Age))
```

## step-2 : Add a column

```python
Names = ['Ramesh','Suresh','Sathis']
Age = [20,25,30]
pd.DataFrame(zip(Names,Age),columns=['Names','Age'])
```

## Step -3 : Add a Index

```python
Names = ['Ramesh','Suresh','Sathis']
Age = [20,25,30]
df1=pd.DataFrame(zip(Names,Age),
             columns=['Names','Age'],
             index=['A','B','C'],)
df1
```

## Step-4 : Overwrite a Names 

```python
df1['Names']=['Sathis','Suresh','Ramesh']
df1
```

```python
# change the index a b c t0 1 2 3
Names = ['Ramesh','Suresh','Sathis']
Age = [20,25,30]
df1=pd.DataFrame(zip(Names,Age),
             columns=['Names','Age'],
             index=[1,2,3])
df1
```

## Step-7 : Drop the column

- before drop the column
- i will create a new column 
- i have only two columns here name and ages 
- i will add city coulumn and then i will drop it 

```python
Names = ['Ramesh','Suresh','Sathis']
Age = [20,25,30]
df1=pd.DataFrame(zip(Names,Age),
             columns=['Names','Age'],
             index=[1,2,3])
df1['City']=['Hybd','Mumbai','Pune']
df1
```

```python
df1.drop('City',axis=1,inplace=True)
```

- axis=1----------------It reference columns 
- axis=0----------------It reference rows
- inplace=True----------IT will overwrite 
- df1.drop('City',axis=1-It will not overwrite

```python
df1
```
