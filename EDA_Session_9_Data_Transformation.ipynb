## Exponential data

```python
exp_data = np.random.exponential(size=1000)
exp_data
```

```python
plt.hist(exp_data,bins=40,label='Exponential')
plt.legend()
plt.show()
```

### Step -2

### Normal distribution data 

```python
norm_data=np.random.normal(size=1000)
plt.hist(norm_data)
```

### Step-3 :

**Log Transformation**

- np.log is used for log transformation 

- generally log transformation will not convert data into normal

- It avoids skew ness

- np.log means natural logarithm base= e 

```python
log_data=np.log(exp_data)
log_data[:10]
```

```python
plt.hist(log_data,bins=40,label='Log Data')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1).hist(exp_data,
                        bins=40,
                        label='Exponential')
plt.legend()
plt.subplot(1,2,2).hist(log_data,
                        bins=40,
                        label='Log Transformation')
plt.legend()
plt.show()
```

## Step-4

### - **Reciprocal Transformation**

- Reciprocal transformation fails when data has zero value 

```python
# take the exp_data
# rec_data = 1/exp_data
# plot histogram 

rec_data= 1 / exp_data
rec_data

```

```python
plt.hist(rec_data,bins=40,label='Reciprocal Data')
plt.legend()
plt.show()
```

## Step-5 

### **Square root transformation**

```python
print(25**2)  # square
print(25**(1/2))  # sqroot

print(np.sqrt(25))
```

```python
sqrt_data=np.sqrt(exp_data)
sqrt_data
```

```python
plt.hist(sqrt_data,bins=40,label='Sqrt Data')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1).hist(exp_data,
                        bins=40,
                        label='Exponential')
plt.legend()
plt.subplot(1,2,2).hist(sqrt_data,
                        bins=40,
                        label='Sqrt Data')
plt.legend()
plt.show()
```

## Step - 6

### Power Transformation

- It is related to sklearn package 

- Package name : sklrearn.preprocessiong

- Method name : Power Transform

- Inside Box-Cox , yeo-jhomson

```python
exp_data = np.random.exponential(size=1000)
exp_data

from sklearn.preprocessing import PowerTransformer

pt=PowerTransformer()

pt_data1=pt.fit([exp_data]).transform([exp_data])
pt_data1

plt.hist(pt_data1[0])
plt.show
```

```python
exp_data = np.random.exponential(size=1000)
exp_data

from sklearn.preprocessing import PowerTransformer

pt=PowerTransformer()

pt_data=pt.fit_transform([exp_data])
pt_data

plt.hist(pt_data[0])
plt.show()
```
