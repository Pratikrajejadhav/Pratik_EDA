## map

- first consider one categoricl column

- observe how many unique labels are there 

- Assume here we are taking case_status column

- case_status two unique labels are there : 
    
    - Denied
    
    - Certified 
    
- Create a dict by providing key:value 
    
- Here the kay= denied  and the value =1

- key=Certified  and the value = 0

- dict will looks like : {'Certified':0,'Denied':1}

```python
visa_df['case_status'].unique()
```

```python
d1={'Certified':0,'Denied':1}
visa_df['case_status']
```

```python
# convert into numerical by map method

d1={'Certified':0,'Denied':1}
visa_df['case_status'].map(d1)


# after mapping , in which column you want apply the data 
# you want to crete a new columns
# you want to use a old columns
```

```python
visa_df['case_status']=visa_df['case_status'].map(d1)
```

```python
visa_df.head()
```

```python
visa_df['continent'].unique()
```

```python
d1={}
labels=visa_df['continent'].unique()
for i in range (len(labels)):
    d1[labels[i]]=i
    
visa_df['continent']=visa_df['continent'].map(d1)
visa_df.head()
```

# Method-3

## $one$-$hot$-$encoder$

**pd.getdummies**

- one hot encoder means at a time only one will be ON,Other is OFF

- ON represent with one , and OFF represent with zero

- One hot encoder creates new extra columns based on labels in the column 

- for example case_status has two unique labels 

- if u apply one hot encoder on case_status it creates two extra columns 

   - case_status_Denied
    
   - case_status_Certified

---------
|Case status||case_status_Denied||case_status_Certified|
|-----------||------------------||---------------------|
|Denied     ||       1          ||          0          |
|Certified  ||       0          ||          1         |

```python
**Advantage**

- One hot encoder gives independense between the variables 

- It is one of the imporantt property before apply models

- Variables are independent of each other

- Variables are perpendicular of each other

- Variables are orthogonal of each other

- Variables are 90degrees phase shift

**Disadvantage**

- Assume that a columns that 100 unique labels

- if we apply one hot encoder it will create 100 new columns 

- your data matrix become sparse 

- we required more time and more memory 
```

```python
# Read the data again

df1="C:\\Users\\admin\\OneDrive\\PRATIK\\NARESH IT\\DATA SCIENCE\\DATA SETS\\csv\\Visadataset.csv"
visa_df=pd.read_csv(df1)
visa_df

# drop the ID before apply one hot encoder 
visa_df.drop('case_id',axis=1,inplace=True)

#apply one hot encoder 
pd.get_dummies(visa_df,dtype='int')
```

```python
#read data
df1="C:\\Users\\admin\\OneDrive\\PRATIK\\NARESH IT\\DATA SCIENCE\\DATA SETS\\csv\\Visadataset.csv"
visa_df=pd.read_csv(df1)
visa_df
print(visa_df['continent'].values[:10])

# Read the Package
from sklearn.preprocessing import LabelEncoder

# Save the Package
le=LabelEncoder()

# Apply fit transform
visa_df['continent']=le.fit_transform(visa_df['continent'])
visa_df.head()

```

```python
#read data
df1="C:\\Users\\admin\\OneDrive\\PRATIK\\NARESH IT\\DATA SCIENCE\\DATA SETS\\csv\\Visadataset.csv"
visa_df=pd.read_csv(df1)
visa_df


# Read the Package
from sklearn.preprocessing import LabelEncoder

# Save the Package
le=LabelEncoder()

# Apply fit transform
# we cant directly apply for all the data 
# do with loop

cat_cols=visa_df.select_dtypes(include='object').columns

for i in cat_cols:
    visa_df[i]=le.fit_transform(visa_df[i])
    
visa_df.head()
```
